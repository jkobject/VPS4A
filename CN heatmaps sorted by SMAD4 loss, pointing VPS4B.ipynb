```python
import pandas as pd
import seaborn as sns
import numpy as np
from math import pi
from bokeh.io import show, output_notebook
from bokeh.models import LinearColorMapper, BasicTicker, PrintfTickFormatter, ColorBar
from bokeh.plotting import figure
from bokeh.models.glyphs import Text, Rect
from bokeh.models import Span
from bokeh.palettes import small_palettes
output_notebook()
```

```python
cn = pd.read_csv('../VPS4A/analysis_scripts/saved.csv', index_col=False)
```

```python
ipos = 61056423
smad4= 48556590
sample_order = []
    
sample_order = cn[cn['Start']<smad4][cn['End']>smad4]["Sample"].values[np.argsort(cn[cn['Start']<smad4][cn['End']>smad4]["Segment_Mean"])]
ordersegmentvalue = cn[cn['Start']<smad4][cn['End']>smad4]["Segment_Mean"].values[np.argsort(cn[cn['Start']<smad4][cn['End']>smad4]["Segment_Mean"])]

orderbyhist = {}
for val in sample_order:
    htype = cn[cn['Sample'] == val].iloc[0]["Histological_Type"]
    if htype in orderbyhist.keys():
        orderbyhist[htype].append(val)
    else:
        orderbyhist.update({htype:[val]})

sample_order = []     
historder = [list(orderbyhist.keys())[i] for i in reversed(np.argsort(np.array([len(val) for val in orderbyhist.values()])))]
for val in historder:
    sample_order.extend(orderbyhist[val])
```

```python
cn['Sample'] = cn['Sample'].astype(str)
df = cn[['Sample','Start','End','Segment_Mean','Histological_Type']]
```

```python
prev=''
counts=[]
flank={'up':50000000,'down':10000000}
for k, val in df.iterrows():
    if val['Start'] >= ipos - flank['up'] and val['Start'] <= ipos + flank['down']: #if start is whitin the window
        if val['End']> ipos + flank['down']:
            val['End'] = ipos + flank['down']
            df.loc[k] = val
    elif val['End'] >= ipos - flank['up'] and val['Start'] < ipos +flank['down']: # if the end is wihtin the window
        val['Start'] = ipos - flank['up']
        if val['End'] > ipos + flank['down']: # if one segment goes through it all 
            val['End'] = ipos+flank['down']
        df.loc[k] = val
    else:
        df = df.drop(k)
        continue
    counts.append(val['End']-val['Start'])

df['size']=counts
```

```python
scale = 100000
df[['Start','End','size']] = np.trunc(df[['Start','End','size']].values / scale).astype(int)
df['Start'] = df['Start'].values + df['size'].values/2
```

```python
# this is the colormap from the original NYTimes plot
colors = ["#75968f", "#a5bab7", "#c9d9d3", "#e2e2e2", "#dfccce", "#ddb7b1", "#cc7878", "#933b41", "#550b1d"]
mapper = LinearColorMapper(palette=colors, low=df.Segment_Mean.min(), high=df.Segment_Mean.max())

TOOLS = "hover,save,pan,box_zoom,reset,wheel_zoom"
p = figure(title="CN heatmaps sorted by SMAD4 loss, pointing VPS4B",
           y_range=(df.End.max(),df.Start.min()),
           x_range=sample_order,
           x_axis_location="above", plot_width=900, plot_height=400,
           tools=TOOLS, toolbar_location='below',
           tooltips=[('pos', '@Start, @End'), ('relative CN', '@Sample')])

p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "5pt"
p.axis.major_label_standoff = 10
p.xaxis.major_label_orientation = pi / 3
pos = 0
#for i,val in enumerate(historder):
#    p.rect(x=pos,y=-7,width=len(orderbyhist[val]), height=10, fill_color=small_palettes['Viridis'][4][i])
#    p.text(x=pos+len(orderbyhist[val])/2, y=-9, text=str(val), text_color="#96deb3")
#    pos += len(orderbyhist[val])
    

p.rect(x="Sample", y="Start", width=0.9, height="size",
       source=df.reset_index(drop=True),
       fill_color={'field': 'Segment_Mean', 'transform': mapper},
       line_color=None)


color_bar = ColorBar(color_mapper=mapper, major_label_text_font_size="5pt",
                     ticker=BasicTicker(desired_num_ticks=len(colors)),
                     formatter=PrintfTickFormatter(format="%.2f"),
                     label_standoff=6, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')
hline = Span(location=609, dimension='width', line_color='green', line_width=0.2)
p.renderers.extend([hline])

show(p)      # show the plot
```
